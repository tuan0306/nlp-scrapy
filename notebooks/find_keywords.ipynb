```python
import json
import re
from collections import Counter
```

```python
file_path='../data/raw/raw_data.json'
with open(file_path,'r',encoding='utf8') as f:
    raw_data=json.load(f)
```

```python
print(f'So luong bai dang: {len(raw_data)}')
```

Gom cac mo ta lai thanh 1 van ban

```python
all_descriptions=" ".join([item.get('description','').lower() for item in raw_data])
```

```python
all_descriptions
```

Loai bo cac dau cau va ky tu dac biet

```python
clean_text=re.sub(r'[^\w\s]','',all_descriptions)
words=clean_text.split()
words
```

Ham ghep tu

```python
def get_ngrams(word_list,n):
    return [" ".join(word_list[i:i+n]) for i in range(len(word_list)-n+1)]
```

```python
bigrams=get_ngrams(words,2)
trigrams=get_ngrams(words,3)
bigrams
```

```python
all_phrases=bigrams+trigrams
phrases_count=Counter(all_phrases)
phrases_count.most_common()
```
